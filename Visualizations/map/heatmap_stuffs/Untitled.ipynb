```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
```

```python
sparql = SPARQLWrapper("http://dbpedia.org/sparql")
sparql.setQuery("""
    SELECT DISTINCT ?city ?totpop
    WHERE { ?city rdf:type dbo:City ; 
        rdfs:label ?label ; 
        dbo:country dbr:Italy;     
        dbo:populationTotal ?totpop
}
""")
```

```python
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
pop_dict={}
#print(results)
for elem in results["results"]["bindings"]:
    city_name=str(elem['city']['value']).replace("http://dbpedia.org/resource/","").replace("Metropolitan_City_of_","").replace("Capital","")
    #print( elem['totpop']['value'])
    pop_dict[city_name]=elem['totpop']['value']
    
pop_dict
```

```python
df=pd.DataFrame.from_dict(pop_dict, orient="index")
```

```python
df.to_csv("city_pop.csv", index=False)
```
