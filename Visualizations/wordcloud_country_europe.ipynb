# List of european country

```python
df_es = df[df['Country']=='es']
df_it = df[df['Country']=='it']
df_pt = df[df['Country']=='pt']
df_se = df[df['Country']=='se']
df_gb = df[df['Country']=='gb']
df_fi = df[df['Country']=='fi']
df_fr = df[df['Country']=='fr']
df_nl = df[df['Country']=='nl']
df_de = df[df['Country']=='de']
```

## Create text for generate wordcloud by country

```python
text_es = ''.join(df_es.Keywords).replace("\'","")
text_it = ''.join(df_it.Keywords).replace("\'","")
text_pt = ''.join(df_pt.Keywords).replace("\'","")
text_se = ''.join(df_se.Keywords).replace("\'","")
text_gb = ''.join(df_gb.Keywords).replace("\'","")
text_fi = ''.join(df_fi.Keywords).replace("\'","")
text_fr = ''.join(df_fr.Keywords).replace("\'","")
text_nl = ''.join(df_nl.Keywords).replace("\'","")
text_de = ''.join(df_de.Keywords).replace("\'","")
print ("There are {} words in espana\n\
There are {} words in italy\n\
There are {} words in portugal\n\
There are {} words in sweden\n\
There are {} words in gran bretain\n\
There are {} words in finland\n\
There are {} words in france\n\
There are {} words in netherlands\n\
There are {} words in germany\n".format(len(text_es),len(text_it),len(text_pt),len(text_se),len(text_gb)
                                       ,len(text_fi),len(text_fr),len(text_nl),len(text_de)))
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=mask_es).generate(text_es)
image_colors = ImageColorGenerator(mask_es)
plt.figure(figsize=[7,7])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
wc.to_file("wordcloud_flag_es.png")
```
