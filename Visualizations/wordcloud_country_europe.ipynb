```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
from PIL import Image
```

```python
df = pd.read_csv('https://media.githubusercontent.com/media/DBertazioli/ARmeetup/master/Visualizations/map/event_topic_keywords.csv')
```

```python
df.head()
```

```python
text = text.replace("\'","")
```

```python
type(text)
```

```python
wordcloud = WordCloud(max_font_size=60, max_words=50, background_color="white").generate(text)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='hamming')
plt.axis("off")
plt.show()
```

# Spain

```python
df_es = df[df['Country']=='es']
df_es.head()
```

```python
text = ''.join(df_es.Keywords)
text = text.replace("\'","")
print ("There are {} words in this country".format(len(text)))
```

```python
mask = np.array(Image.open("spain.png"))
wordcloud_spa = WordCloud(background_color="white", max_words=1000, mask=mask).generate(text)
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wordcloud_spa.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
spagna_mask = np.array(Image.open("spagna.png"))
spagna_mask
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=spagna_mask)
image_colors = ImageColorGenerator(mask)
# Generate a wordcloud
wc.generate(text)

# store to file
#wc.to_file("img/wine.png")

# show
plt.figure(figsize=[12,6])
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis("off")
plt.show()
```
